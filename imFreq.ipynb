## Import Data
In order to ease our analysis of contrasts we opt to convert the image to grayscale on import.

```python
from skimage import io
example_file_path = "http://i128.photobucket.com/albums/p185/Inventor0001/checkers.jpg" 
#you can replace our example_file_path with a link to your image of choice
image = io.imread(example_file_path, as_grey=True)
```

## Frequency Spectrum Extraction
This is one of the tricky bits of what imFreq does. Here we will apply the `freq_processing` function defined in the `imfreq.py` file shipped with our script suite.

`freq_processing` takes a 2d array as an input, and consecutively:

* Performs `np.fft.fft2` on it.
* Takes the absoulte values of the `fft2` output.
* Iteratively calculates the orientation-agnostic frequency for the amplitude at each position.

```python
from imfreq import freq_extraction
freq_spectrum = freq_extraction(image)
```

## Sorting and Grouping
The frequency spectrum thus obtained is not ordered.
The following function:

* orders the output so that frequency values are incremntal.
* sums amplitudes if their corresponding frequency values are identical.

```python
from imfreq import freq_sort
freq_spectrum_sorted = freq_sort(freq_spectrum)
```

The dataframe containing the frequency data, `freq_spectrum_sorted`, now looks like this:

```python
print(freq_spectrum_sorted[::3*10**4]) #print every 30.000th value for demonstrative purposes
```

## Frequency Spectrum Plot
Now we can plot our progress thus far - using a log scale (since the 0/pixel frequency band - the DC offset - and bands close to it have large values in images with a non-zero mean).

```python
from imfreq import plot_dots
#plot_dots only processes lists of dataframes with the same index and column names as our example above
plot_dots(freq_spectrum_sorted, image=image)
```

## Compare Frequency Spectra

So now that we know what the frequency spectrum of an image looks like, we can try comparing multiple images.

```python
input_list = [
              "http://31.media.tumblr.com/1b891da6d4391810c5aea0828b7339fb/tumblr_mzob0uRPrn1sjalwio1_1280.jpg",
              "http://25.media.tumblr.com/e08947b6b08a42ef95b6450d593e5f06/tumblr_mzq2kvDbsY1sjalwio1_1280.jpg",
              "http://25.media.tumblr.com/7f07082c4788772336b18e9a47423547/tumblr_mzraqkhk9Q1sjalwio1_1280.jpg",
              "http://i128.photobucket.com/albums/p185/Inventor0001/checkers.jpg"
              ]
for item in input_list:
    print(item)
    image = io.imread(item, as_grey=True)
    freq_spectrum = freq_extraction(image)
    freq_spectrum_sorted = freq_sort(freq_spectrum)
    plot_dots(freq_spectrum_sorted, image=image)
    print("done")              
```
